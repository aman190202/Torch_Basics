```python
import torch
import torch.nn as nn
```

```python
X = torch.tensor([[1],[2],[3],[4]],dtype=torch.float32)
Y = torch.tensor([[100],[200],[300],[400]],dtype=torch.float32)

X_test=torch.tensor([5],dtype=torch.float32)

n_samples,n_features=X.size()

print(n_samples,n_features)

input_size=n_features
output_size=n_features

model=nn.Linear(input_size,output_size)
```

```python
print(f'Prediction f(5) before training : {model(X_test).item():.3f}')
learning_rate=0.01
n_iters=1000
loss=nn.MSELoss()
optimizer=torch.optim.SGD(model.parameters(),lr=learning_rate)
for epoch in range(n_iters):
    y_pred=model(X)
    l=loss(Y,y_pred)

    l.backward()
    optimizer.step()
    optimizer.zero_grad()

    [w,b]=model.parameters()
    print(f'Epoch {epoch+1}, w : {w[0][0].item():.3f} , loss : {l:.3f}')

print(f'Prediction f(5) after training : {model(X_test).item():.3f}')
```
